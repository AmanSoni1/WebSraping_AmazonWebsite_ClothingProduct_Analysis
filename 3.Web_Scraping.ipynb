***IMPORTING LIBRARIES***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot
import requests
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
```

***Requesting URL***

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.amazon.in/s?i=apparel&bbn=29945467031&rh=n%3A1571271031%2Cn%3A1953602031%2Cn%3A11400137031%2Cn%3A29945467031%2Cp_n_specials_match%3A21618256031%2Cp_85%3A10440599031%2Cp_n_feature_nineteen_browse-bin%3A11301357031&s=apparel&dc&_encoding=UTF8&content-id=amzn1.sym.796255b6-64ab-41b0-b0a5-cc30823f832b&pd_rd_r=86987011-6bfe-4081-8825-5f6ca58aaa93&pd_rd_w=1pLJZ&pd_rd_wg=g23Xv&pf_rd_p=796255b6-64ab-41b0-b0a5-cc30823f832b&pf_rd_r=PYW5BHQC318ZJHEAYW99&qid=1694848142&rnid=11301356031&ref=sr_pg_5',headers=headers).text
```

```python

soup=BeautifulSoup(webpage,'lxml')
soup


print(soup.prettify())
```

### Scraping Product Names from Amazon Website Page 1

```python
# Product name:
# <span class="a-size-base-plus a-color-base">FUNDAY FASHION</span>
product_names=[]
for i in soup.find_all('span',class_='a-size-base-plus a-color-base a-text-normal'):
  print(i.text.strip())
  product_names.append(i.text.strip())
    
product_names
```

```python
len(product_names)
```

### Scraping Product Prices from Amazon Website Page 1

```python
# Price
# <span class="a-price-whole">378</span>


product_price=[]
for i in soup.find_all('span',class_='a-price-whole'):
  print(i.text.strip())
  product_price.append(i.text.strip())
    
product_price
```

```python
len(product_price)
```

### Scraping Product Original Prices from Amazon Website Page 1

```python
# Original Price
# <span class="a-offscreen">₹1,499</span>

product_orgprice=[]
for i in soup.find_all('span',class_='a-offscreen'):
  print(i.text.strip())
  product_orgprice.append(i.text.strip()) 

    
product_orgprice=product_orgprice[1::2]
product_orgprice
```

```python
len(product_orgprice)
```

### Scraping Product Reviews from Amazon Website Page 1

```python
# Review
# <span class="a-size-base s-underline-text">(160)</span>
product_reviews=[]
for i in soup.find_all('span',class_='a-size-base s-underline-text'):
  print(i.text.strip())
  product_reviews.append(i.text.strip()) 

    
len(product_reviews)
```


### Scraping Product Ratings from Amazon Website Page 1

```python
#Product ratings
#Ratings
# <span class="a-size-base puis-bold-weight-text">3.7</span>

P_ratings=[]
for i in soup.find_all("span",class_="a-size-base puis-bold-weight-text"):
  print(i.text.strip())
  P_ratings.append(i.text.strip()) 

    
P_ratings
```

### Combining all the Scraped List into a single DataFrame Shown Below:

### Product_Names
### Product_Ratings
### Product_Price
### Product_ORGPrice
### Product_Reviews

```python
df=pd.DataFrame({"Product_Names":product_names,"Product_Ratings":P_ratings,"Product_Price":product_price,"Product_ORGPrice":product_orgprice,
               "Product_Reviews":product_reviews})
df
```

```python
### Saving dataframe As a csv file
df.to_csv("Amazon_Clothing_Page5.csv",index=False)
```

### Combining all these info from 5 consecutive pages of the Amazon Website

```python


file_paths = ['Amazon_Clothing_Page1.csv', 'Amazon_Clothing_Page2.csv', 'Amazon_Clothing_Page3.csv','Amazon_Clothing_Page4.csv','Amazon_Clothing_Page5.csv']

combined_data = pd.DataFrame()
for file_path in file_paths:
    df = pd.read_csv(file_path)
    combined_data = pd.concat([combined_data, df], ignore_index=True)
combined_data
```

```python
combined_data.sort_values(by=["Product_Ratings"],ascending=False)
```
